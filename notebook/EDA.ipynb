<h2>Graphical view of data set.</h2>

<p>The given data set have numerical values and these value are vary in a range. So we simply try to create a hist plot to check the data distribution.</p>

```python
## we use matplotlib and seaborn to plot the data set.

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')



next_plot=1

## By using `for` loop to plot as hist plot's

for i in df.columns:
    fig=plt.figure(figsize=(30, 200))  ## defining the size of figure
    plt.subplot(12,1,next_plot)    
    sns.histplot( df[i])   ## creating the histplot for every column.
 
    plt.show()
    
    next_plot += 1
    
```

<h4>From the above plots we can see the value of each columns are vary in a range. And the target column `quality` it is very from 3 to 8,. It is a multi-classification problem set.</h4>

<h4>If we take a look on quality column plot it is clear the maximum data output is 5 and 6, and have maximum value counts. One the other hand another output like 3,4,7,8 are not much populated as 3 and 5. So if we try to make model from this dataset it will be a dumb model. Now we have to populate this imbalance dataset. </h4>

```python
from imblearn.over_sampling import SMOTE 
from sklearn.model_selection import train_test_split
import pandas as pd 


x=df.iloc[:,:-1]
y=df['quality']

x_train,  x_test, y_train, y_test=train_test_split(x,y, random_state=42, test_size=0.30)

# Apply SMOTE to balance the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train, )

new_df=pd.concat([X_train_resampled, y_train_resampled], axis=1)

y_train_resampled.value_counts()
```

<h4>Now the data is balance.</h4>

```python
import numpy as np

### Here, we try to round the values of oversampled data as raw data set have.

new_df[['fixed acidity', 'residual sugar', 'free sulfur dioxide', 'total sulfur dioxide', 'alcohol']]=np.round(new_df[['fixed acidity', 'residual sugar', 'free sulfur dioxide', 'total sulfur dioxide', 'alcohol']],1)
new_df[['volatile acidity','citric acid', 'pH', 'sulphates']]=np.round(new_df[['volatile acidity', 'citric acid', 'pH', 'sulphates']],2)
new_df['chlorides']=np.round(new_df['chlorides'], 3)
new_df['density']=np.round(new_df['density'], 6)

new_df.to_csv('raw_data/Raw_Data_Red_Wine.csv', index=False)
```

<h3>Creating the classification models and selecting model with higher accuracy.</h3>

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd



x=new_df.iloc[:,:-1]
y=new_df['quality']

x_train,  x_test, y_train, y_test=train_test_split(x,y, random_state=42, test_size=0.30)


models=[BernoulliNB(), RandomForestClassifier(),  SVC(), DecisionTreeClassifier()]   ## list of some classifiaction models.

for i in models:
    
   model=i
   model.fit(X_train_resampled, y_train_resampled)
   y_pred=model.predict(x_test)
   
   print(f'MOdel: {str(i)}')
   print('Model accuracy: ',accuracy_score(y_true=y_test, y_pred=y_pred))
   print('Classification report: ', '\n', classification_report(y_true=y_test, y_pred=y_pred))
   print('==========================================================')
   print('\n')
   
```

<h5>from the above observation we can see that the RandomForestClassifier got the higher accuracy which is 1.0, But to reduce the model overfiting we use hyper perimeter tunning for this model. </h5>

<h2>Hyperperimeter tunning</h2>

```python
from sklearn.model_selection import GridSearchCV
import warnings 
warnings.filterwarnings('ignore')


RFC=RandomForestClassifier()  ## Selected model.

## perimeter dict to try.
param={
    "n_estimators": [100,  300, 400, 700],
    "criterion":['gini', 'entropy', 'log_loss'],
    "max_depth":[3,4,5,6],
    "max_features": ['sqrt', 'log2', None],
    "oob_score":[ True, False]
   
   
}
grc=GridSearchCV(RFC, param_grid=param, cv=3, n_jobs=10)
grc.fit(x_test, y_test)
```

```python
grc.best_params_  ## best perimeters for our model.
```

```python
## Testing new model accuracy

from sklearn.metrics import classification_report, accuracy_score

y2_pred=grc.predict(x_test)

print('Model accuracy: ',accuracy_score(y_true=y_test, y_pred=y2_pred))
print('\n')

print('classification_report :', '\n',classification_report(y_pred=y2_pred, y_true=y_test))
```
